```python
import pandas as pd

df_repu = pd.read_csv(
    "../data/interim/df_repu.csv", low_memory=False, dtype={"ID_orateur": str}
)
df_annot = pd.read_csv(
    "../data/interim/annotations_republique-test-base_default_train.csv",
    low_memory=False,
    dtype={"ID_orateur": str},
)
```

```python
df_joined = df_repu.merge(
    df_annot,
    left_on="ID_paragraphe",
    right_on="dataset_ID_paragraphe",
    how="left",
    suffixes=("_repu", "_annot"),
)
df_joined.head()
```

```python
df_joined = df_joined.drop(
    columns=[
        "Unnamed: 0",
        "id",
        "dataset_ID_paragraphe",
        "text",
        "user",
        "timestamp",
        "comment",
    ]
)
```

```python
df_joined = df_joined.rename(columns={"default": "previous_annotation"})
```

```python
import csv

df_joined.to_csv(
    "../data/interim/df_first_annotation.csv", index=False, quoting=csv.QUOTE_ALL
)
```
