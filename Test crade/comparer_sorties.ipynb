```python
import pandas as pd
```

```python
df_v3 = pd.read_csv("V3.csv")
print(df_v3.shape)

df_v3.head()
```

```python
df_v4 = pd.read_csv("V4.csv")
print(df_v4.shape)
df_v4.head()
```

```python
df_complet_v3 = pd.read_csv("export_v3.csv", low_memory=False)
```

```python
df_complet_v4 = pd.read_csv("df_complet.csv", low_memory=False)
```

```python
# Trouver les id_paragraphe présents dans l'un mais pas dans l'autre pour df_complet_v3 et df_complet_v4
ids_test = set(df_complet_v3["ID_paragraphe"])
ids_full = set(df_complet_v4["ID_paragraphe"])

ids_in_test_not_in_full = ids_test - ids_full
ids_in_full_not_in_test = ids_full - ids_test

# Afficher les lignes concernées pour analyse
diff_test = df_complet_v3[df_complet_v3["ID_paragraphe"].isin(ids_in_test_not_in_full)]
diff_full = df_complet_v4[df_complet_v4["ID_paragraphe"].isin(ids_in_full_not_in_test)]

print("Dans df_complet_v3 mais pas dans df_complet_v4 :")
display(diff_test)

print("Dans df_complet_v4 mais pas dans df_complet_v3 :")
display(diff_full)
```
